```python
from itertools import product, permutations, combinations, combinations_with_replacement
```

image.png

```python
from os import pread
lst = 'ABCD'
pd_result = product(lst, repeat = 2)
for r1 in pd_result:
    print(r1)
```

```python
# 순열
per_result = permutations(lst, 2)
for r2 in per_result:
    print(r2)
```

```python
# 조합
comb_result = combinations(lst, 2)
for r3 in comb_result:
    print(r3)
```

```python
# 중복 있는 조합
rep_comb = combinations_with_replacement(lst, 2)
for r4 in rep_comb:
    print(r4)
```
